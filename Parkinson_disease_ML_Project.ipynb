```python
#import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("parkinsons.data")

#display first few rows to verify
df.head()
```

```python
#to check structure and data types
df.info()

#summarize statistics
df.describe()

#check for missing values
df.isnull().sum()
```

```python
#Data Processing
#delete the non-numeric attribute i.e., name

df = df.drop("name", axis=1, errors='ignore')
X = df.drop("status", axis=1)
y = df["status"]
```

```python
#feature scaling

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#train_test_split()
#divide our parkinsons data into features (X) and labels (y).

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
#Model training Logistic Regression model for binary classification

from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python

# Predict on test data

y_pred = model.predict(X_test)

# model performance evaluation

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
# Visualize confusion matrix

plt.figure(figsize=(6, 4))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```
